# Kohonen Network

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def gauss_nhood(x, t=None):
    if t is None:
        t = 1
    return np.exp(-(x * t) ** 2)

def mexican_hat_nhood(x, t=None):
    if t is None:
        t = 1
    return (2 - 4*(x*t) ** 2) * np.exp(-(x * t) ** 2)

def exp_decay(t, lambda_max):
    """ t: current interation """
    return np.exp(-t / lambda_max)

def euclidean_distance(a, b):
    return np.linalg.norm(np.subtract(a, b), ord=2, axis=-1)

def standardize_data(X):
    """
    returns:
    X_new - standardized X
    a tuple (mean, std) - normal distribution parameters from X for destandarizing
    """
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    X_new = (X - mean) / std
    return X_new, (mean, std)
```

```python
class KOH:

    def __init__(self, input_dim, shape, distance_metric=euclidean_distance, decay_function=exp_decay,
                 neighborhood_function=gauss_nhood, initialization='net'):
        """
        input_dim: number of input vectors, neuron dim
        shape: tuple (N, M)
        """
        assert len(shape) == 2
        self.input_dim = input_dim
        self.shape = shape
        self.weights = {'random': self.random_init(),
                        'net': self.net_init()}[initialization]
        self.distance_metric = distance_metric
        self.decay_function = decay_function
        self.neighborhood_function = neighborhood_function
        self.data = None
    
    def random_init(self):
        return np.random.standard_normal(
            size=(self.shape[0], self.shape[1], self.input_dim)
        )
    
    def net_init(self):
        x = np.linspace(-2,2,self.shape[1])
        y = np.linspace(-2,2,self.shape[0])
        x_net, y_net = np.meshgrid(x,y)
        return np.dstack((x_net, y_net))
    
    def nhood_weight(self, n1, n2, t=None, r=None):
        """ calculates neighborhood distance """
        if r is None:
            r = 1
        return self.neighborhood_function(self.distance_metric(n1, n2) * r, t)

    def find_bmu(self, x):
        """ finds best matching unit """
        index = np.argmin(self.distance_metric(self.weights, x))
        return np.unravel_index(index, self.shape)
    
    def get_weights(self, indices):
        """ get weights from indices """
        return self.weights[indices[0]][indices[1]]
                    
    def update_weights(self, x, bmu, t, lambda_max, r, use_nhood_decay=False):
        """ updates weights """
        time = None
        if use_nhood_decay:
            time = t
            
        nhood_weights = self.nhood_weight(bmu, self.weights, time, r)
        nhood_weights_tensor = np.stack((nhood_weights, ) * self.input_dim, axis=2)
        
        self.weights = self.weights + exp_decay(t, lambda_max) * nhood_weights_tensor * (x - self.weights)

    def predict(self, data):
        predictions = []
        for x in data:
            predictions.append(np.argmin(self.distance_metric(self.weights, x)))
        return np.array(predictions)
    
    def train(self, data, lambda_max, r=None, use_nhood_decay=False):
        """ 
        lambda_max: number of iterations
        """
        self.data = data
        shuffled_data = np.copy(data)
        np.random.shuffle(shuffled_data)
        
        for t in range(1, lambda_max+1):
            for x in shuffled_data:
                bmu_indices = self.find_bmu(x)
                bmu = self.get_weights(bmu_indices)
                self.update_weights(x, bmu, t, lambda_max, r, use_nhood_decay)

    def plot_clusters(self, labels, type='hex', return_plot=False):
        """ labels: pandas series """
        if type == 'hex':
            weights_res = self.weights.reshape(-1, 2)
            sns.scatterplot(x=self.data[:,0], y=self.data[:,1], hue=labels.to_numpy())
            sns.scatterplot(x=weights_res[:,0], y=weights_res[:,1])
        elif type == 'cube':
            weights_res = self.weights.reshape(-1, 3)
            fig = plt.figure()
            ax = fig.add_subplot(111, projection='3d')
            ax.scatter(self.data[:,0], self.data[:,1], self.data[:,2], c=labels.to_numpy(), marker='o', alpha=0.2)
            ax.set_xlabel('X')
            ax.set_ylabel('Y')
            ax.set_zlabel('Z')
            ax.scatter(weights_res[:,0], weights_res[:,1], weights_res[:,2], c='k', alpha=1)
        if return_plot:
            return plt
        else:
            plt.show()
```

## Data

```python
hexagon = pd.read_csv("dane/hexagon.csv")
cube = pd.read_csv("dane/cube.csv")
```

```python
hex_train = hexagon[['x', 'y']].to_numpy()
cube_train = cube[['x', 'y', 'z']].to_numpy()
```

```python
sns.scatterplot(x=hex_train[:,0], y=hex_train[:,1], hue=hexagon['c'].to_numpy()).set_title('Hexagon dataset')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(cube_train[:,0], cube_train[:,1], cube_train[:,2], c=cube['c'].to_numpy(), marker='o')  # Color is blue, marker is circle
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Cube dataset')
plt.show()
```

## Tests

```python
# K = M x N
# C - number of classes
```

```python
from sklearn.metrics import v_measure_score
```

```python
koh = KOH(2, (4, 6))
koh.data = hex_train
koh.plot_clusters(labels=hexagon['c'])
```

### Hexagon dataset

```python
# K < C
```

```python
koh = KOH(2, (3, 1))
```

```python
koh.train(hex_train, lambda_max=10)
```

```python
preds = koh.predict(hex_train)
print('clusters:', len(np.unique(preds)))
print('v measure:', v_measure_score(hexagon['c'].to_numpy(), preds))
```

```python
koh.plot_clusters(labels=hexagon['c'])
```

Uywanie liczby K mniejszej ni liczba klas jest bez sensu, gdy nie mamy szans wykrycia wszystkich klastrw. V-measure wynosi 0.67

```python
# K = C
```

```python
koh = KOH(2, (3, 2))
```

```python
koh.train(hex_train, lambda_max=10)
```

```python
preds = koh.predict(hex_train)
print('clusters:', len(np.unique(preds)))
print('v measure:', v_measure_score(hexagon['c'].to_numpy(), preds))
```

```python
koh.plot_clusters(labels=hexagon['c'])
```

Uycie K = C ma najwikszy sens, gdy wtedy kady neuron odpowiada jednemu klastrowi

```python
# K > C
```

```python
koh = KOH(2, (30, 20))
```

```python
koh.train(hex_train, lambda_max=10)
```

```python
preds = koh.predict(hex_train)
print('clusters:', len(np.unique(preds)))
print('v measure:', v_measure_score(hexagon['c'].to_numpy(), preds))
```

```python
koh.plot_clusters(labels=hexagon['c'])
```

Widzimy e przy uyciu bardzo gstej siatki dostajemy wietne wyniki. Ju przy 10 iteracjach punkty siatki zbiegaj do 6 konkretnych punktw. Normalnie nie znamy prawdziwej iloci klastrw, dlatego warto stosowa o wiele gstsz siatk.

Sprawdmy, jak dziaa parametr r

```python
plt.figure(figsize=(12,4))
r_values = [0.5, 1, 1.5]
for i, r in enumerate(r_values):
    koh = KOH(2, (30, 20))
    koh.train(hex_train, lambda_max=5, r=r)
    preds = koh.predict(hex_train)
    plt.subplot(1, 3, i+1)
    plt.title(f"r={r}, clusters: {len(np.unique(preds))}, v-score: {round(v_measure_score(hexagon['c'].to_numpy(), preds), 2)}")
    koh.plot_clusters(labels=hexagon['c'], return_plot=True)
```

```python
koh = KOH(2, (30, 20), neighborhood_function=mexican_hat_nhood)
koh.train(hex_train, lambda_max=100)
preds = koh.predict(hex_train)
print('clusters:', len(np.unique(preds)))
print('v measure:', v_measure_score(hexagon['c'].to_numpy(), preds))
koh.plot_clusters(labels=hexagon['c'])
```

Zdecydowanie lepiej, uzyskalimy v-score na poziomie 0.87, jednake nadal nie jest to idealna klasteryzacja. Przy uyciu mniejszej lambdy udao si uzyska jeszcze lepszy v-measure i 8 klastrw.

```python
koh = KOH(3, (8, 6))
koh.train(cube_train, lambda_max=40, r=1.2, use_nhood_decay=True)
preds = koh.predict(cube_train)
print('clusters:', len(np.unique(preds)))
print('v measure:', v_measure_score(cube['c'].to_numpy(), preds))
koh.plot_clusters(labels=cube['c'], type='cube')
```

Przy przeprowadzonych eksperymentach wida, e zwykle liczba klastrw jest mniejsza ni oczekiwana. Nie ma zatem sensu uywanie wikszych wartoci lambda i mniejszych wartoci parametru r, gdy powoduj one zmniejszenie liczby klastrw. eby jednak sprawiedliwoci stao si zado, sprawdmy co si stanie przy uyciu duej wartoci lambda np. 1000:
